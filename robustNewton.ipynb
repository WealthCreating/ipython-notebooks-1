##2- Método de Newton-Raphson ORIGINAL com Jacobiano sintético - apenas vetorial e loops explícitos

```python
#          Por E R Woiski Ilha Solteira - SP    #
from numpy import *

def newtonRaphson(fun,x0,jacob = None, nitermax = 1000, xtol=1.e-8,args=None):
    """
Implementação direta do método original de Newton-Raphson para sistema de equações
não lineares. fun é a função vetorial de x (cada x >=o) e de args, que são argumentos
suplementares, x0 são as condições iniciais e jacob é a função jacobiana obtida
algébricamente ou numericamente por diferenças finitas, se jacob is None.
    """
    error = 2*xtol
    ite = 0
    x = asarray(x0)
     
    while error > xtol and ite <= nitermax:
        F = fun(x,args)

        ''' se o jacobiano não for fornecido,
            calcule um por diferenças finitas...'''
        if jacob is None:
            y = x.copy()
            linhas = colunas = len(x)
            J = zeros((linhas,colunas))
            
            # A definição correta de derivadas parciais mas HORRÍVEL DUPLO LAÇO!
            for i in range(linhas): 
                for j in range(colunas):
                    y[j] = x[j]*(1. + xtol)
                    J[i,j] = (fun(y,args) - fun(x,args))[i]/(x[j]*xtol)
                    y = x.copy()
        else:
            J = jacob(x,args)
                    
        # seja qual for a forma de determinação do jacobiano...    
        dx = solve(J,-F)
    
        ite += 1
        x += dx
        F = fun(x,args)
        # O erro é o maior valor dentre a norma quadrática e o maior resíduo...
        error = max(dot(F,F),abs(F).max())

    if ite == nitermax:
        print u'não convergiu com %s iterações!' %nitermax
        
    return x,ite,F
```

##3- Método de Newton-Raphson ROBUSTO com Jacobiano sintético - vetorial E ESCALAR - loops explícitos

```python
#          Por E R Woiski Ilha Solteira - SP    #
from numpy import *

def robustNewton_local(fun,x0,jacob = None, nitermax = 1000, xtol=1.e-8,args=None):
    """
Implementação direta do método de Newton-Raphson tornada robusta pelo
controle do avanço de cada componente do vetor dx, para manter cada x
sempre positivo. fun é a função vetorial de x (cada x >=o), x0 são as
condições iniciais e jacob é a função jacobiana obtida algébricamente
ou numericamente por diferenças finitas, se jacob is None.
    """
    error = 2*xtol
    ite = 0
    x = asarray(x0)
    
    if size(x0) != 1: #  VETORIAL!...
        x = asarray(x0)
        linhas = colunas = size(x0)
        F = fun(x, args)
     
        while error > xtol and ite <= nitermax:

            ''' se o jacobiano não for fornecido,
                calcule um por diferenças finitas...'''
            if jacob is None:
                y = x.copy()
                J = zeros((linhas,colunas))
                
                # NOTE que com o slice em J, o laço em i desapareceu....
                for j in range(colunas):
                    y[j] = x[j]*(1. + xtol)
                    J[:,j] = (fun(y,args) - fun(x,args))/(x[j]*xtol)
                    y = x.copy()
            else:
                J = jacob(x,args)
                    
            # seja qual for a forma de determinação do jacobiano...    
            dx = solve(J,-F)

            # se algum componente de xi + dxi for negativo, reduza dxi pela metade...               
            for i in range(len(x)):
                #while abs(fun(x+dx,args))>abs(F): dx[i] *=.5
                while -dx[i] > x[i]: dx[i] *= .5
                 
            ite += 1
            x += dx
            F = fun(x,args)
            # O erro é o maior valor dentre a norma quadrática e o maior resíduo...
            error = max(dot(F,F),abs(F).max())
        
    else: #ESCALAR!...
        x = x0
        F = fun(x, args)
        
        while error > xtol and ite <= nitermax:
                                                                               
            if jacob is None:
                J = (fun(x*(1.+xtol),args) - F)/(x*xtol)
                ''' se o jacobiano não for fornecido,
                    calcule um por diferenças finitas.'''
                                                  
            else: # ou então jacob fornecido...
                J = jacob(x,args)
                        
            # seja qual for a forma de determinação do jacobiano...    
            dx = -F/J

            # se x + dx for negativo, reduza dx pela metade...
            #while abs(fun(x+dx,args)) > abs(F): dx *= .5
            while -dx > x: dx *= .5

            ite += 1
            x += dx # avança x...
            F = fun(x, args)
                
            error = abs(F)

    if ite == nitermax:
        print u'não convergiu com %s iterações!' %nitermax
        
    return x,ite,F
```

##4- Método de Newton-Raphson robusto com Jacobiano sintético - vetorial e escalar - FORMA VETORIZADA

```python
#          Por E R Woiski Ilha Solteira - SP    #
from numpy import array,dot,size,asarray,identity,abs #ones,copy,zeros
from numpy import apply_along_axis as apaxis
from numpy.linalg import solve

def robustNewton_vetor(fun,x0,jacob = None, nitermax = 200, xtol=1.e-8,args=None):
    """
Implementação direta do método de Newton-Raphson tornada robusta pelo
controle do avanço de cada componente do vetor dx, para manter cada x
sempre positivo. fun é a função vetorial de x (cada x >=o), x0 são as
condições iniciais e jacob é a função jacobiana obtida algébricamente
ou numericamente por diferenças finitas, se jacob is None.
            ** VERSÃO DIDÁTICA 12/02/2013 **
    """
    error = 2*xtol
    ite = 0
    errors = []
    
    if jacob is None:
        print "Sem Jacobiano..."
    else:
        print "Com Jacobiano..."
    
    if size(x0) != 1: #  VETORIAL!...
        x = asarray(x0)
        linhas = colunas = size(x0)
        ixtol = identity((colunas))*xtol
        F = fun(x, args)
                    
        while error > xtol and ite <= nitermax:

            if jacob is None:
                ''' se o jacobiano não for fornecido,
                calcule um por diferenças finitas...'''
                # NOTE: Nenhum laço,mas veja a explicação lá no fim.
         
                J = (apaxis(fun,1,x*(ixtol + 1.),args)-F).transpose()/(x*xtol)
                #J = (array(map(lambda x:fun(x,args),x*(ixtol + 1.))) -
                  #F ).transpose()/(x*xtol)
            else:
                J = jacob(x,args)
                    
            # seja qual for a forma de determinação do jacobiano...    
            dx = solve(J,-F)
            
            # se deltaFi for positivo,reduza dxi pela metade...
            # se algum componente de xi + dxi for negativo, reduza dxi pela metade...
            for i in range(linhas):
                while (abs(fun(x+dx,args)) - abs(F))[i] > 0.: dx[i] *= .5
            #while dx[(x+dx)<0].size != 0:
                #dx[(x+dx)<0] *= .5
                 
            ite += 1
            x += dx
            F = fun(x,args)
            
            # O erro é o maior valor dentre a norma quadrática e o maior resíduo...
            error = max(abs(F).max(),dx.max())
            
            errors.append(error)
            print ite,x,F,error
        
    else: #ESCALAR!...
        x = x0
        F = fun(x, args)
        
        while error > xtol and ite <= nitermax:
                                                                               
            if jacob is None:
                J = ((fun(x*(1.+xtol),args) - F)/(x*xtol) or
                     (fun(x*(1.-xtol),args) - F)/(-x*xtol) or
                     (fun(x*(1.+xtol),args) - fun(x*(1.-xtol),args))/(2*x*xtol) )
                ''' se o jacobiano não for fornecido,
                    calcule um por diferenças finitas.'''
                                                  
            else: # ou então jacob fornecido...
                J = jacob(x,args)
                        
            # seja qual for a forma de determinação do jacobiano...    
            dx = -F/J
            
            # se deltaF for positivo, reduza dx pela metade...
            # se x + dx for negativo, reduza dx pela metade...
            while abs(fun(x+dx,args)) > abs(F): dx *= .5
            #while -dx > x: dx *= .5

            ite += 1
            x += dx # avança x...
            F = fun(x, args)
                
            error = max(abs(F),abs(dx))
            
            errors.append(error)
            print ite,x,F,error
            
    semilogy(range(1,ite+1),errors,'go-')    # plote a convergência....
    xlabel(u'n° de interações')
    ylabel('error')
    grid()
    show()
    
    print '\n','*'*80,'\n'
    
    if ite == nitermax:
        print u'não convergiu com %s iterações!' %nitermax
        
    return x,ite,F
```
